```python
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
```

```python
# Sample dataset
corpus = [
    "Cats and Dogs are not allowed.",
    "Cats and Dogs are antagonistic.",
    "Birds fly high in the sky.",
    "The quick brown fox jumps over the lazy dog.",
    "Never underestimate the power of a good book.",
    "A journey of a thousand miles begins with a single step.",
    "Coding is fun and rewarding.",
    "Learning new things is essential for growth.",
    "Stay curious and keep exploring.",
    "The world is full of wonders waiting to be discovered."
]

```

```python
# Create a CountVectorizer to transform the text into BOW
vectorizer = CountVectorizer()

# Fit and transform the corpus
X_bow = vectorizer.fit_transform(corpus)

# Convert to an array for simplicity
X_bow = X_bow.toarray()
X_bow
```

```python
vectorizer_unigram = CountVectorizer(ngram_range=(1, 1))
X_unigram = vectorizer_unigram.fit_transform(corpus)
print("Unigrams:", vectorizer_unigram.get_feature_names_out())

```

```python
# Bigrams
vectorizer_bigram = CountVectorizer(ngram_range=(2, 2))
X_bigram = vectorizer_bigram.fit_transform(corpus)
print("Bigrams:", vectorizer_bigram.get_feature_names_out())

```

```python
# Trigrams
vectorizer_trigram = CountVectorizer(ngram_range=(3, 3))
X_trigram = vectorizer_trigram.fit_transform(corpus)
print("Trigrams:", vectorizer_trigram.get_feature_names_out())
```

```python
# prompt: Based on unigram, bigram and trigram do some visualisation

import matplotlib.pyplot as plt
import seaborn as sns

# Unigram Visualization
unigram_counts = X_unigram.sum(axis=0)
unigram_freq = dict(zip(vectorizer_unigram.get_feature_names_out(), unigram_counts.tolist()[0]))

plt.figure(figsize=(10, 5))
sns.barplot(x=list(unigram_freq.keys()), y=list(unigram_freq.values()))
plt.title('Unigram Frequencies')
plt.xlabel('Unigrams')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout() # Adjust the plot to ensure labels are not cut off
plt.show()


# Bigram Visualization
bigram_counts = X_bigram.sum(axis=0)
bigram_freq = dict(zip(vectorizer_bigram.get_feature_names_out(), bigram_counts.tolist()[0]))

plt.figure(figsize=(10, 5))
sns.barplot(x=list(bigram_freq.keys()), y=list(bigram_freq.values()))
plt.title('Bigram Frequencies')
plt.xlabel('Bigrams')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()


# Trigram Visualization
trigram_counts = X_trigram.sum(axis=0)
trigram_freq = dict(zip(vectorizer_trigram.get_feature_names_out(), trigram_counts.tolist()[0]))

plt.figure(figsize=(10, 5))
sns.barplot(x=list(trigram_freq.keys()), y=list(trigram_freq.values()))
plt.title('Trigram Frequencies')
plt.xlabel('Trigrams')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
